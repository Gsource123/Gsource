### Data Prep
We now use the `Llama-3.1` format for conversation style finetunes. We use [Maxime Labonne's FineTome-100k](https://huggingface.co/datasets/mlabonne/FineTome-100k) dataset in ShareGPT style. But we convert it to HuggingFace's normal multiturn format `("role", "content")` instead of `("from", "value")`/ Llama-3 renders multi turn conversations like below:

```
<|begin_of_text|><|start_header_id|>user<|end_header_id|>

Hello!<|eot_id|><|start_header_id|>assistant<|end_header_id|>

Hey there! How are you?<|eot_id|><|start_header_id|>user<|end_header_id|>

I'm great thanks!<|eot_id|>
```

We use our `get_chat_template` function to get the correct chat template. We support `zephyr, chatml, mistral, llama, alpaca, vicuna, vicuna_old, phi3, llama3` and more.

````python
# <<< START OF YOUR CUSTOM DATA PREPARATION CELL (for ProjectID-PartID Naming Convention) >>>
import os
import re
from collections import defaultdict # For grouping TXT files
from datasets import Dataset
from google.colab import drive
import gc
import torch # Assuming torch is imported for torch.cuda.empty_cache()

# Mount Google Drive
try:
    drive.mount('/content/drive', force_remount=True)
    print("Google Drive mounted successfully.")
except Exception as e:
    print(f"Error mounting Google Drive: {e}")
    raise

# Define Directories for your data on Google Drive
TXT_FILES_DIR = '/content/drive/MyDrive/Batch_1/Points/' # Your TXT files directory
DXF_FILES_DIR = '/content/drive/MyDrive/Batch_1/DXF/'   # Your DXF files directory

def parse_txt_filename_for_project_and_part(filename):
    """
    Parses filenames like 'ANY_PREFIX_ENDING_WITH_PROJECT_ID-PART_ID.txt'.
    The PROJECT_ID is considered everything before the last hyphen-number sequence.
    Returns (PROJECT_ID_STRING, PART_ID_AS_INT) or (None, None).
    Example: "MyProject-Survey A-01.txt" -> ("MyProject-Survey A", 1)
             "123-Data Set Alpha-2.txt" -> ("123-Data Set Alpha", 2)
    """
    base_name = os.path.splitext(filename)[0]
    # Regex breakdown:
    # ^          - Start of the string
    # (.*)       - Group 1 (Project ID): Captures any characters (non-greedy, followed by a hyphen)
    #            - It will capture everything up to the *last* occurrence of "-<digits>"
    #   -        - Matches a literal hyphen
    #   (\d+)    - Group 2 (Part ID): Captures one or more digits
    # $          - End of the string
    match = re.match(r'^(.*)-(\d+)$', base_name)
    if match:
        project_id_str = match.group(1).strip() # .strip() to remove potential trailing spaces in project_id
        try:
            part_id_int = int(match.group(2))
            if not project_id_str: # Handle cases like "-1.txt" where project_id might be empty
                return None, None
            return project_id_str, part_id_int
        except ValueError:
            # This should not happen if \d+ matches, but good for safety
            return None, None
    return None, None

def create_dataset_from_project_parts(txt_base_dir, dxf_base_dir, max_projects_to_load=None):
    data_for_dataset = []
    projects_processed_count = 0
    skipped_projects_log = []

    if not os.path.isdir(txt_base_dir):
        print(f"ERROR: TXT directory not found: {txt_base_dir}")
        return data_for_dataset, skipped_projects_log
    if not os.path.isdir(dxf_base_dir):
        print(f"ERROR: DXF directory not found: {dxf_base_dir}")
        return data_for_dataset, skipped_projects_log

    project_txt_files_map = defaultdict(list)
    all_txt_filenames = [f for f in os.listdir(txt_base_dir) if f.endswith(".txt")]

    print(f"Scanning {len(all_txt_filenames)} .txt files to group by project ID...")
    for txt_filename in all_txt_filenames:
        project_id, part_id = parse_txt_filename_for_project_and_part(txt_filename)
        if project_id is not None and part_id is not None:
            project_txt_files_map[project_id].append({'filename': txt_filename, 'part_id': part_id, 'full_path': os.path.join(txt_base_dir, txt_filename)})
        else:
            skipped_projects_log.append(f"TXT file '{txt_filename}' did not match expected 'PROJECT_ID-PART_ID.txt' format or had issues parsing.")

    print(f"Found {len(project_txt_files_map)} unique project IDs from TXT filenames.")

    for project_id, parts_info_list in sorted(project_txt_files_map.items()):
        if max_projects_to_load is not None and projects_processed_count >= max_projects_to_load:
            print(f"DEBUG: Reached processing limit of {max_projects_to_load} projects.")
            break

        sorted_parts = sorted(parts_info_list, key=lambda x: x['part_id'])

        # Expected DXF filename is ProjectID.dxf (where ProjectID is what was extracted from TXT)
        expected_dxf_filename = f"{project_id}.dxf"
        dxf_filepath = os.path.join(dxf_base_dir, expected_dxf_filename)

        if not os.path.exists(dxf_filepath):
            skipped_projects_log.append(f"Project '{project_id}': Skipped (DXF file '{expected_dxf_filename}' not found at '{dxf_filepath}').")
            continue

        combined_txt_content_list = []
        all_parts_read_successfully = True
        for part_info in sorted_parts:
            try:
                with open(part_info['full_path'], 'r', encoding='utf-8') as f_txt:
                    combined_txt_content_list.append(f_txt.read().strip())
            except Exception as e:
                skipped_projects_log.append(f"Project '{project_id}', part '{part_info['filename']}': Error reading TXT file: {e}")
                all_parts_read_successfully = False
                break

        if not all_parts_read_successfully or not combined_txt_content_list:
            skipped_projects_log.append(f"Project '{project_id}': Skipped (Failed to read all parts or no content found in parts).")
            continue

        try:
            with open(dxf_filepath, 'r', encoding='utf-8', errors='ignore') as f_dxf:
                dxf_content = f_dxf.read().strip()
        except Exception as e:
            skipped_projects_log.append(f"Project '{project_id}': Error reading DXF file '{expected_dxf_filename}': {e}")
            continue

        final_txt_input = "\n\n--- Next Point File Part ---\n\n".join(combined_txt_content_list) # Customize separator if needed

        if final_txt_input and dxf_content:
            instruction_prompt = "The following point data is combined from multiple sequentially numbered input files for a single project. Convert this combined survey point data into a single, complete, and valid DXF file format. Ensure the DXF includes all necessary header, table, and entity sections for all the points provided."
            formatted_text = f"<s>[INST] {instruction_prompt}\n\nCombined Project Point Data (Parts sorted by number):\n{final_txt_input} [/INST]\nDXF Output:\n{dxf_content}</s>"
            data_for_dataset.append({"text": formatted_text})
            projects_processed_count += 1
        else:
            skipped_projects_log.append(f"Project '{project_id}': Skipped (Empty combined TXT or DXF content after processing).")

    print(f"Successfully processed {projects_processed_count} projects into {len(data_for_dataset)} training examples.")
    if skipped_projects_log:
        print(f"INFO: Encountered issues or skipped {len(skipped_projects_log)} projects/files during processing:")
        for info in skipped_projects_log[:20]: print(f"  - {info}")
        if len(skipped_projects_log) > 20: print(f"  ... and {len(skipped_projects_log)-20} more.")

    return data_for_dataset

# --- Main part of data prep cell ---
# This 'max_seq_length' should ideally be defined in the cell where you load the model.
# It's used by the SFTTrainer later.
if 'max_seq_length' not in globals():
    print("WARNING: 'max_seq_length' not defined globally. Setting a default for safety (e.g., 1024 or 2048). Ensure it matches model config.")
    max_seq_length = 2048 # Adjust this to match your model's capability and DXF lengths

# Number of distinct projects to load. Each project will become one training example.
NUMBER_OF_PROJECTS_TO_LOAD = 5 # <<<<<<< START WITH A VERY SMALL NUMBER (e.g., 1, 2, 5)

print(f"Attempting to load data for up to {NUMBER_OF_PROJECTS_TO_LOAD} projects...")
list_of_data_dicts = create_dataset_from_project_parts(
    TXT_FILES_DIR,
    DXF_FILES_DIR,
    max_projects_to_load=NUMBER_OF_PROJECTS_TO_LOAD
)

if not list_of_data_dicts:
    if NUMBER_OF_PROJECTS_TO_LOAD < 10 and len(list_of_data_dicts) == 0:
         print(f"Loaded 0 examples with a limit of {NUMBER_OF_PROJECTS_TO_LOAD} projects. This might be okay if your first few projects had parsing/matching issues or were skipped.")
         dataset = Dataset.from_list([]) # Create empty dataset to allow pipeline to continue for debugging
    else:
        # If you expect data and get none with a larger limit, it's an error.
        raise ValueError("No training data was loaded based on project parts. Please check your file naming convention, paths, and content. Review skipped_projects_log for details.")
else:
    print(f"Collected {len(list_of_data_dicts)} data items (projects). Converting to Hugging Face Dataset...")
    dataset = Dataset.from_list(list_of_data_dicts)

# Clean up memory
del list_of_data_dicts
gc.collect()
if 'torch' in globals() and torch.cuda.is_available():
    torch.cuda.empty_cache()
print("Memory cleanup attempted after list_of_data_dicts deletion.")

if len(dataset) > 0:
    print("\nSample processed data point (first 500 chars of the 'text' field):")
    print(dataset[0]['text'][:500] + "...")
    print(f"\nTotal training examples (projects) in dataset: {len(dataset)}")
else:
    print("\nNo data points in the final dataset.")
# <<< END OF YOUR CUSTOM DATA PREPARATION CELL (for ProjectID-PartID Naming Convention) >>>
````

### Train the model
Now let's use Huggingface TRL's `SFTTrainer`! More docs here: [TRL SFT docs](https://huggingface.co/docs/trl/sft_trainer). We do 60 steps to speed things up, but you can set `num_train_epochs=1` for a full run, and turn off `max_steps=None`. We also support TRL's `DPOTrainer`!

````python
# <<< START OF YOUR CUSTOM TRAINING CELL >>>
import torch # <--- ADD THIS LINE
from trl import SFTTrainer
from transformers import TrainingArguments
# from unsloth import is_bfloat16_supported # is_bfloat16_supported is now torch.cuda.is_bf16_supported()

if not 'dataset' in globals() or len(dataset) == 0:
    raise RuntimeError("The 'dataset' variable is not defined or is empty. Please run the data preparation cell first.")

# Define Training Arguments for your DXF task
# Adjust these based on your dataset size and available GPU
training_args = TrainingArguments(
    per_device_train_batch_size = 1,  # DXF files can be very long, start with 1
    gradient_accumulation_steps = 2,  # Effective batch size = 1 * 8 = 8. Adjust as needed.
    warmup_steps = 10,                 # Number of steps for learning rate warmup
    # max_steps = 100,                 # Option 1: Train for a fixed number of steps. Good for small datasets.
                                      # (~39 examples / effective batch 8 = ~5 steps per epoch)
                                      # So 100 steps = ~20 epochs.
    num_train_epochs = 15,            # Option 2: Train for a fixed number of epochs. (e.g., 10-20 for ~39 examples)
                                      # If both max_steps and num_train_epochs are set, max_steps takes precedence.
                                      # Choose one or the other, or set max_steps high if using num_train_epochs.
    learning_rate = 1e-4,             # A slightly smaller LR for LoRA might be good for complex structured data (1e-4 to 2e-4)
    fp16 = not torch.cuda.is_bf16_supported(), # Use FP16 if BF16 not supported (standard for T4)
    bf16 = torch.cuda.is_bf16_supported(),     # Use BF16 if supported (A100, H100)
    logging_steps = 1,                # How often to log training loss
    optim = "adamw_8bit",             # Unsloth recommends adamw_8bit for 4-bit models
    weight_decay = 0.01,
    lr_scheduler_type = "linear",     # Learning rate scheduler
    seed = 3407,
    output_dir = "./unsloth_finetuned_dxf_model", # Directory to save checkpoints and final model
    save_strategy = "epoch",          # Save a checkpoint at the end of each epoch
    # report_to = "tensorboard",      # Optional: for logging to TensorBoard or WandB
)

# Create the SFTTrainer
trainer = SFTTrainer(
    model = model,                     # Your Unsloth-prepared Llama model
    tokenizer = tokenizer,             # The Llama tokenizer
    train_dataset = dataset,           # YOUR CUSTOM DATASET (loaded in the previous cell)
    dataset_text_field = "text",       # The field in your dataset containing the formatted string
    max_seq_length = max_seq_length,   # Max sequence length defined earlier
    dataset_num_proc = 2,              # Number of processes for dataset mapping (can speed it up)
    packing = False,                   # Set to False for long sequences like DXF. True is for packing short sequences.
    args = training_args,
)
# <<< END OF YOUR CUSTOM TRAINING CELL >>>
````

We also use Unsloth's `train_on_completions` method to only train on the assistant outputs and ignore the loss on the user's inputs.

### GGUF / llama.cpp Conversion
To save to `GGUF` / `llama.cpp`, we support it natively now! We clone `llama.cpp` and we default save it to `q8_0`. We allow all methods like `q4_k_m`. Use `save_pretrained_gguf` for local saving and `push_to_hub_gguf` for uploading to HF.

Some supported quant methods (full list on our [Wiki page](https://github.com/unslothai/unsloth/wiki#gguf-quantization-options)):
* `q8_0` - Fast conversion. High resource use, but generally acceptable.
* `q4_k_m` - Recommended. Uses Q6_K for half of the attention.wv and feed_forward.w2 tensors, else Q4_K.
* `q5_k_m` - Recommended. Uses Q6_K for half of the attention.wv and feed_forward.w2 tensors, else Q5_K.

[**NEW**] To finetune and auto export to Ollama, try our [Ollama notebook](https://colab.research.google.com/github/unslothai/notebooks/blob/main/nb/Llama3_(8B)-Ollama.ipynb)

````python
# Save to 8bit Q8_0
if False: model.save_pretrained_gguf("model", tokenizer,)
# Remember to go to https://huggingface.co/settings/tokens for a token!
# And change hf to your username!
if False: model.push_to_hub_gguf("hf/model", tokenizer, token = "")

# Save to 16bit GGUF
if False: model.save_pretrained_gguf("model", tokenizer, quantization_method = "f16")
if False: model.push_to_hub_gguf("hf/model", tokenizer, quantization_method = "f16", token = "")

# Save to q4_k_m GGUF
if False: model.save_pretrained_gguf("model", tokenizer, quantization_method = "q4_k_m")
if False: model.push_to_hub_gguf("hf/model", tokenizer, quantization_method = "q4_k_m", token = "")

# Save to multiple GGUF options - much faster if you want multiple!
if False:
    model.push_to_hub_gguf(
        "hf/model", # Change hf to your username!
        tokenizer,
        quantization_method = ["q4_k_m", "q8_0", "q5_k_m",],
        token = "", # Get a token at https://huggingface.co/settings/tokens
    )
````
